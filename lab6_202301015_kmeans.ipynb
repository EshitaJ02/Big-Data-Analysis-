```python
# Find a dataset in the form of (x,y) and then apply K-Mean clustering and find
# number of Clusters using Elbow method.

import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import pandas as pd

df=pd.read_csv("https://raw.githubusercontent.com/AnjulaMehto/MCA/main/Mall_Customers.csv")
print(df)
```

```python
print(df.columns[3],df.columns[4])
data=list(zip(df['Annual Income (k$)'],df['Spending Score (1-100)']))
kmeans = KMeans(n_clusters=5, max_iter=50)
kmeans.fit(data)
```

```python
KMeans(max_iter=50, n_clusters=5)
```

```python
plt.scatter(df['Annual Income (k$)'], df['Spending Score (1-100)'], c=kmeans.labels_)
plt.show()
```
